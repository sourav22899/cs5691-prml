```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm
```

```python
rate = pd.read_csv('ratings.csv')
rem = pd.read_csv('remarks.csv')
rem_sup = pd.read_csv('remarks_supp_opp.csv')
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
rate = rate.loc[rate['emp'] >= 0]
rem = rem.loc[rem['emp'] >= 0]
rem_sup = rem_sup.loc[rem_sup['emp'] >= 0]
rem = rem.dropna()
rem_sup = rem_sup.dropna()
rate = rate.dropna()
rem = rem.drop_duplicates()
rem_sup = rem_sup.drop_duplicates()
rate = rate.drop_duplicates()
```

```python
emps = rate['emp'].unique()
comps = rate['comp'].unique()
```

```python
rate_list = []
for comp in tqdm(comps):
    comp_wise_list = rate.loc[rate['comp'] == comp]
    for emp in emps:
        r = comp_wise_list.loc[rate['emp'] == emp]
        if r.shape[0]:
            rate_list.append([emp,comp,list(r['rating']),r['rating'].shape[0]])
```

```python
# rate[(rate['emp'] == 16) & (rate['comp'] == 'bnivzbfi')]
rate
```

```python
rate_df = pd.DataFrame(rate_list,columns=['emp','comp','rating','freq'])
```

```python
rate_df
```

```python
train_rate_list = []
# train_emps = train['emp'].unique()
# train_comps = train['comp'].unique()
for _,row in train.iterrows():
    train_rate_list.extend(list(rate_df[(rate_df['emp'] == row['emp']) & (rate_df['comp'] == row['comp'])]['rating']))
```

```python
train['ratings'] = train_rate_list
train
```

```python
remarks = rem.remarkId.unique()
from collections import Counter
remark_wt_list = []
for remark in remarks:
    rel_list = rem_sup[rem_sup['remarkId'] == remark]
    remark_wt_list.append([remark,rel_list.shape[0],Counter(rel_list['support'])[True]])
```
